```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_json("JSON_USAHA_PERKBLI.json")
df.head()
```

```python
df["nama_indikator"] = df["nama_indikator"].str.replace(
    r'Kategori KBLI\s*"([A-Z])"',  # cari pola: Kategori KBLI "X"
    r"\1",  # ganti dengan huruf X saja
    regex=True,
)
df["total_value"] = df["total_value"].fillna(0).astype(int)

df["id_wilayah_7"] = df["id_wilayah"].astype(str).str[:7]
```

```python
data_list = df["nama_indikator"].unique().tolist()
data_list
```

```python
# 2. Filter dulu baris "Jumlah Usaha" (kalau memang mau khusus itu saja)
usaha_per_kbli = df[
    df["nama_indikator"].str.contains("Jumlah Usaha", case=False, na=False)
]

# 3. Pivot: nama_indikator jadi kolom, isinya dari total_value
usaha_per_kbli_pivot = usaha_per_kbli.pivot_table(
    index=["nama_kecamatan", "id_wilayah_7"],
    columns="nama_indikator",
    values="total_value",
    aggfunc="first",  # atau 'sum' kalau ada duplikat dan mau dijumlah
).reset_index()

kolom_numerik = usaha_per_kbli_pivot.select_dtypes(include='number').columns
total_row = usaha_per_kbli_pivot[kolom_numerik].sum()
total_row['nama_kecamatan'] = 'Total'
total_row['id_wilayah_7'] = ''

# Gabungkan ke bawah
usaha_per_kbli_pivot = pd.concat(
    [usaha_per_kbli_pivot, total_row.to_frame().T],
    ignore_index=True
)

usaha_per_kbli_pivot
```

```python
omzet_per_kbli = df[
    df["nama_indikator"].str.contains("Total Omzet", case=False, na=False)
]

omzet_per_kbli_pivot = omzet_per_kbli.pivot_table(
    index=["nama_kecamatan", "id_wilayah_7"],
    columns="nama_indikator",
    values="total_value",
    aggfunc="first",
).reset_index()

# --- Hitung ulang total_row khusus untuk omzet ---
kolom_numerik = omzet_per_kbli_pivot.select_dtypes(include="number").columns
total_row = omzet_per_kbli_pivot[kolom_numerik].sum()
total_row["nama_kecamatan"] = "Total"
total_row["id_wilayah_7"] = ""

# Gabungkan ke bawah
omzet_per_kbli_pivot = pd.concat(
    [omzet_per_kbli_pivot, total_row.to_frame().T], ignore_index=True
)

omzet_per_kbli_pivot
```

```python
nilai_tambah_per_kbli = df[
    df["nama_indikator"].str.contains("Nilai Tambah", case=False, na=False)
]

nilai_tambah_per_kbli_pivot = nilai_tambah_per_kbli.pivot_table(
    index=["nama_kecamatan", "id_wilayah_7"],
    columns="nama_indikator",
    values="total_value",
    aggfunc="first",  # atau 'sum' kalau ada duplikat dan mau dijumlah
).reset_index()
# --- Hitung ulang total_row khusus untuk omzet ---
kolom_numerik = nilai_tambah_per_kbli_pivot.select_dtypes(include="number").columns
total_row = nilai_tambah_per_kbli_pivot[kolom_numerik].sum()
total_row["nama_kecamatan"] = "Total"
total_row["id_wilayah_7"] = ""

# Gabungkan ke bawah
nilai_tambah_per_kbli_pivot = pd.concat(
    [nilai_tambah_per_kbli_pivot, total_row.to_frame().T], ignore_index=True
)
nilai_tambah_per_kbli_pivot
```

```python
sheets = {
    "Jumlah Usaha": usaha_per_kbli_pivot,
    "Total Omzet": omzet_per_kbli_pivot,
    "Nilai Tambah": nilai_tambah_per_kbli_pivot,
}
tanggal_hari_ini = datetime.now().strftime("%d-%m-%Y")
with pd.ExcelWriter(f"rekap_perkbli_{tanggal_hari_ini}.xlsx", engine="openpyxl") as writer:
    for sheet_name, df_sheet in sheets.items():
        df_sheet.to_excel(writer, sheet_name=sheet_name, index=False)
```
