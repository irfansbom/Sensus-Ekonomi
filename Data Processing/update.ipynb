```python
total = len(df)

cursor = conn.cursor()

for i, (_, row) in enumerate(df.iterrows(), start=1):

    jumlah_kk = int(row["Keluarga"])
    jumlah_usaha = int(row["UMK"]) + int(row["UM"]) + int(row["UB"])
    jumlah_muatan = int(row["TOTAL ASSIGNMENT FASIH"])

    cursor.execute(
        """
        UPDATE subsls
        SET
            jumlah_kk = %s,
            jumlah_usaha = %s,
            jumlah_muatan = %s
        WHERE id_subsls = %s
        """,
        (
            jumlah_kk,
            jumlah_usaha,
            jumlah_muatan,
            row["IDSUBSLS_25_2"],
        ),
    )

    if i % 100 == 0:
        conn.commit()
        print(f"Progress: {i}/{total}")

conn.commit()
print(f"Selesai: {total}/{total}")
```
