```python
import pandas as pd
```

```python
df = pd.read_excel("../rekap_anomali_by subsls.xlsx")
df["Grand Total"] = pd.to_numeric(df["Grand Total"], errors="coerce").fillna(0).astype(int)
df["idsls"] = df["idsls"].astype(str)
df["idkec"] = df["idsls"].str[:7]
df.head()
```

```python
df_top_10_kecamatan = (
    df.sort_values(["idkec", "Grand Total"], ascending=[True, False])
    .groupby("idkec", group_keys=False)
    .head(10)
)
```

```python
df_top_10_kecamatan.to_excel("../top10_anomali_perkecamatan.xlsx")
```

```python
df = pd.read_excel("../rekap_status_usaha_by_subsls.xlsx")

df["id_wilayah"] = df["id_wilayah"].astype(str)
df["idkec"] = df["id_wilayah"].str[:7]
df["jumlah_tidak_ditemukan"] = df["Jumlah Usaha Tidak Ditemukan (BKU)"].add(
    df["Jumlah Usaha Tidak Ditemukan (Usaha Keluarga)"], fill_value=0
)

df_sorted = df.sort_values(by="jumlah_tidak_ditemukan", ascending=False)
df_top10_kecamatan = df_sorted.groupby("idkec", group_keys=False).head(10)
```

```python
df_top10_kecamatan.to_excel("../top10_usaha_tidak_ditemukan_perkec.xlsx")
```
