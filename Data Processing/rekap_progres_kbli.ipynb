```python
import pandas as pd
from pathlib import Path
```

```python
folder = Path("../scrap_progres_kbli/")
files = sorted(folder.glob("progres_kbli_*.xlsx"))

total_files = len(files)
list_df = []
total_baris = 0

for i, file in enumerate(files, start=1):
    df_temp = pd.read_excel(file)
    list_df.append(df_temp)

    total_baris += len(df_temp)
    print(f"[{i}/{total_files}] {file.name} -> {len(df_temp)} baris (total sejauh ini: {total_baris})")

df = pd.concat(list_df, ignore_index=True)

print(f"\nTotal file : {len(list_df)}")
print(f"Total baris: {len(df)}")
```

```python
df_jumlah_usaha = df[df["nama_indikator"].isin(indikator_jumlah_usaha)]
print(f"Total baris cocok: {len(df_jumlah_usaha)}")
df_jumlah_usaha_subsls = (
    df_jumlah_usaha.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "jumlah_usaha"})
)
df_jumlah_usaha_subsls
```

```python
df_nilai_tambah = df[df["nama_indikator"].isin(indikator_nilai_tambah)]
print(f"Total baris cocok: {len(df_nilai_tambah)}")
df_nilai_tambah_subsls = (
    df_nilai_tambah.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "nilai_tambah"})
)
df_nilai_tambah_subsls
```

```python
df_total_omset = df[df["nama_indikator"].isin(indikator_total_omset)]
print(f"Total baris cocok: {len(df_total_omset)}")
df_total_omset_subsls = (
    df_total_omset.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "total_omset"})
)
df_total_omset_subsls
```

```python
df_total_output = df[df["nama_indikator"].isin(indikator_total_output)]
print(f"Total baris cocok: {len(df_total_output)}")
df_total_output_subsls = (
    df_total_output.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "total_output"})
)
df_total_output_subsls
```

```python
df_tenaga_kerja = df[df["nama_indikator"].isin(indikator_tenaga_kerja)]
print(f"Total baris cocok: {len(df_tenaga_kerja)}")
df_tenaga_kerja_subsls = (
    df_tenaga_kerja.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "tenaga_kerja"})
)
df_tenaga_kerja_subsls
```

```python
df_total_upah = df[df["nama_indikator"].isin(indikator_total_upah)]
print(f"Total baris cocok: {len(df_total_upah)}")
df_total_upah_subsls = (
    df_total_upah.groupby(["id_wilayah", "nama_wilayah"], as_index=False)[
        "total_value"
    ]
    .sum()
    .rename(columns={"total_value": "total_upah"})
)
df_total_upah_subsls
```

```python
df_gabung = (
    df_jumlah_usaha_subsls.merge(df_nilai_tambah_subsls, on=["id_wilayah", "nama_wilayah"], how="outer")
    .merge(df_total_omset_subsls, on=["id_wilayah", "nama_wilayah"], how="outer")
    .merge(df_total_output_subsls, on=["id_wilayah", "nama_wilayah"], how="outer")
    .merge(df_tenaga_kerja_subsls, on=["id_wilayah", "nama_wilayah"], how="outer")
    .merge(df_total_upah_subsls, on=["id_wilayah", "nama_wilayah"], how="outer")
)
```

```python
df_gabung.to_excel("../Rekap Progress Usaha/rekap_nilai_usaha.xlsx")
```
