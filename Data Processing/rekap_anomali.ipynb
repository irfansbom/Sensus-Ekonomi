```python
import glob
import os
import pandas as pd
```

```python
pola_file = r"./SQL Lab/*sumsel_au3*.csv"
daftar_file = glob.glob(pola_file)
if daftar_file:
    # Mengurutkan file berdasarkan waktu modifikasi terbaru
    file_terbaru = max(daftar_file, key=os.path.getmtime)
    # 3. Baca file terbaru tersebut ke dalam DataFrame
    df_au3 = pd.read_csv(file_terbaru, dtype=str)
    print(f"Berhasil membaca file terbaru: {file_terbaru}")
else:
    print("Tidak ada file yang cocok dengan pola tersebut di folder 'SQL Lab'.")
```
