```python
from pathlib import Path
import pandas as pd
```

```python
# Folder tempat file CSV
folder = Path("./raw missing value")

# Ambil semua file yang diawali sqllab_missing dan berakhiran .csv
files = folder.glob("sqllab_sumsel_missing_nik_kosong_*.csv")

# Baca semua file
list_df = [pd.read_csv(file) for file in files]

# Gabungkan menjadi satu DataFrame
df = pd.concat(list_df, ignore_index=True)

print(f"Total file: {len(list_df)}")
print(f"Total baris: {len(df)}")
```

```python
df.to_excel("raw missing value/missing_nik_sumsel.xlsx",index=False)
```
